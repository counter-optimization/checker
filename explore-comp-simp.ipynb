```python
# helper functions
def remove_rows(orig, exclude):
    new = pandas.merge(
      orig, exclude,
      on=["address", "expression"],
      how="outer",
      indicator=True
    )

    new = new.loc[
        new["_merge"] == "left_only"
    ].drop("_merge", axis=1)

    return new
```
