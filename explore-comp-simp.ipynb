```python
"""
Get all records that look like push or pop a quadword:
Add64(..., 0x0000000000000008)
Sub64(..., 0x0000000000000008)
"""
push_or_pop_data = not_simplifiable[
    ((not_simplifiable["expression"].str.contains("Sub64") |
        not_simplifiable["expression"].str.contains("Add64"))
      &
      (not_simplifiable["expression"].str.contains("0x0000000000000008")))
]

def remove_rows(orig, exclude):
    new = pandas.merge(
      orig, exclude,
      on=["address", "expression"],
      how="outer",
      indicator=True
    )

    new = new.loc[
        new["_merge"] == "left_only"
    ].drop("_merge", axis=1)

    return new

not_push_or_pop_data = remove_rows(not_simplifiable, push_or_pop_data)

num_push_or_pop_data, _ = push_or_pop_data.shape
num_not_push_or_pop_data, _ = not_push_or_pop_data.shape
print("There are %s push_or_pop expressions" % num_push_or_pop_data)
print("There are %s not_push_or_pop expressions" % num_not_push_or_pop_data)
assert(num_not_push_or_pop_data + num_push_or_pop_data == num_not_simplifiable)

push_or_pop_data
```
